```python
import requests
import numpy as np 
import pandas as pd
from bs4 import BeautifulSoup
link="https://fr.wikipedia.org/wiki/Naruto"
page=requests.get(link)
soup = BeautifulSoup(page.content, 'html.parser')
```

```python
f = soup.find("h1",{"class":"firstHeading mw-first-heading"})
f.text
```

```python
container =soup.find("div",{"class" :"infobox_v3 noarchive large"})
tables = container.find_all("table")
d = {}
for table in tables:
    trs = table.find("tbody").find_all("tr")
    for tr in trs:
        d[tr.find("th").text] = tr.find("td").text     
d
```

```python
links = soup.find_all("a")
for link in links:
    print(link["href"])
```

```python
def wraps(link):
    page=requests.get(link)
    soup = BeautifulSoup(page.content, 'html.parser')
    f = soup.find("h1",{"class":"firstHeading mw-first-heading"})
    print(f.text)
    container =soup.find("div",{"class" :"infobox_v3 noarchive large"})
    tables = container.find_all("table")
    d = {}
    for table in tables:
        trs = table.find("tbody").find_all("tr")
        for tr in trs:
            d[tr.find("th").text] = tr.find("td").text     
    print("\n")
    print(d)
    links = soup.find_all("a")
    for link in links:
        print(link["href"])

    
wraps("https://fr.wikipedia.org/wiki/Naruto")
```
